### What is Data Visualization ?

Data visualization is the presentation of data in a pictorial or graphical format. It describes any effort to help people understand their data better by placing it in a visual context. Patterns, trends and correlations that would might go undetected without it, can be recognized easier with data visualization. Data visualization presents analytics visually and enables you to grasp difficult concepts or identify new patterns. There is also interactive visualization, that enables you to drill down into charts and graphs for more detail, by interactively changing what you see.

 

### Why do we need Data Visualization ?

The human brain can process information a lot better when it is in a graphical form. A graphical representation of your data gives you a very good understanding about your dataset in general and what features are useful for your machine learning model and which ones not. One of the primary goals of data visualization is to communicate information clearly and efficiently with statistical graphics, plots and information graphics. For example, if you are working at a company and have to present your Data Analysis to your boss, he won’t understand why you made certain assumptions if you don’t show it to him in an easy understandable graphical form. So in order to understand your data in an efficient and quick way, you need Data Visualization.

 

### How does it work ?

First you visualize a dataset to understand what the data actually is telling you. The next step is to make some more detailed analysis. Then you document your assumptions from the two previous steps. Thirdly, you may want to transform your dataset because your assumptions about the data make some features unnecessary, so you want to remove them. Then you visualize your data again and repeat the whole process till you are satisfied.

# How to use it

There are various types of plots available within Matplotlib. Today we will go through the most common ones.

# How to plot a simple graph

```python
# import matplotlib
from matplotlib import pyplot as plt

x = [1,2,3] # x-axis
y = [4,5,1] # y-axis

# plot graphs to the canvas
plt.plot(x, y)

# show our plot
plt.show()
```

Now we will plot the same graph again but now we will label our axes and add a title to our graph.

```python
x = [1,2,3] # x-axis
y = [4,5,1] # y-axis

plt.title("most simple graph") # define the title

plt.ylabel('y') # define the name of the y axis
plt.xlabel("x") # define the name of the x axis 

# plot graphs to the canvas
plt.plot(x, y)

plt.show()
```

# How to style your graph

This time we will plot 2 lines on a graph and style them. 

```python
from matplotlib import style

x = [1,2,3] # x-axis of the first line
y = [4,5,1] # y-axis of the first line

x2 = [4,5,1] # x-axis of the second line
y2 = [1,2,3] # y-axis of the second line

# After we put in the variabels we want to plot (x and y), 
# we specify a color, give the lines a label and 
# specify the line width 
plt.plot(x, y, color ='g', label = 'styled line 1', 
         linewidth = 4)
plt.plot(y,x, color = 'k', label = 'styled line 2', 
         linewidth = 3)

plt.title('styled graph')
plt.xlabel('x')
plt.ylabel('y')

# add a legend to the graph which displays the label
plt.legend()

# adding a white grid to the graph
plt.grid(True, color = 'w')

plt.show()
```

## 1. Bar Graph

A Bar Graph is used to compare things between different groups. They work very good if you try to measure the change of these realtionships over time, when these changes are relatively large. 

```python
x = [2, 4, 6, 8, 10]
y = [7, 7, 3, 7, 9]

x2 = [1, 3, 5, 7, 9]
y2 = [6, 3, 9, 8, 10]

# The only difference to the previous examples 
# is that we use the "plt.bar()" function.
plt.bar(x,y, color='y', label = 'styled line 1')
plt.bar(x2,y2, color='g', label = 'styled line 2')

plt.title('Bar Graph')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```

# 2. Histogram

In Histograms we quantitative variables and bar graphs have categorical variables.

For example if you want to plot the population for different countries, you should use a bar graph because it is categorical and the countries names would therefore be the categories.

A example where you should plot histograms would be if you have a dataset of different age groups and you want to visualize how much each age group is contributing to the gdp of a country.

```python
ages = [0, 22, 33, 42, 21, 20, 44, 59, 63, 19, 41, 73, 
        93, 87, 82, 44, 33, 33, 27, 64, 14, 11, 25, 49, 
        41, 37, 26, 19, 22]

# Specifying the number of bars. The ages we specified 
# above, are now sorted into 10 categories.
number_of_bins = 10

# Here we use the "plt.hist()" function. We tell 
# matplotlib that the histtype should be bars and 
# with rwidth we specify the width of the bars.
plt.hist(ages, number_of_bins, label = 'age/gdp', 
         histtype = 'bar', rwidth = 0.7 )

plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('GDP')

plt.legend()

plt.show()
```

# 3. Scatterplot

We use scatterplots to find out how much 2 variables (for 2 dimensions) or 3 variables (for 3 dimensions) are related to each other (e.g the correlation).

```python
x = [2, 4, 6, 8, 10, 3, 7, 9]
y = [7, 7, 3,6, 8, 10, 7, 9]

# Here we use the "plt.scatter()" function.
plt.scatter(x, y, label = 'Scatterplot', color = 'r')

plt.title('Scatterplot')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```

# 4. Stack Plot 

Stack plots are used to identify changes over time for several groups or categories. They are very similar to the line graphs we discussed earlier in this post. You could for example display the stock market value over the last years and could easily spot the points where the market value started to decline or to rise.

```python
category1 = [2, 2, 3, 4, 5]
category2 = [7, 8, 6, 11, 7]
category3 = [2, 5, 4, 3, 2]
category4 = [7, 8, 7, 2, 2]
category5 = [8, 5, 6, 8, 13]

# The problem here is that we don't have labels for our data 
# and therefore we needed to plot some empty lines and use 
# the "plt.plot" function.
plt.plot([], [], color = 'c', label = 'category1', linewidth = 5)
plt.plot([], [], color = 'k', label = 'category2', linewidth = 5)
plt.plot([], [], color = 'r', label = 'category3', linewidth = 5)
plt.plot([], [], color = 'm', label = 'category4', linewidth = 5)

# Now we build a stackplot out of these empty lines, 
# with the same labels and colors, using "plt.stackplot()".
plt.stackplot(category1, category2, category3, category4, 
              category5, colors=['c', 'k', 'r', 'm'])


plt.title('Stackplot')
plt.xlabel('x')
plt.ylabel('y')

#plt.legend()

plt.show()
```

# 5. Pie Chart

Pie Chart are very similar. They are very good to display things like market share of 
companies or something like this.

```python
# Matplotlib handles the sizes of the slices, 
# so we just have to put in some numbers.
slices = [14, 2, 2, 4, 3]

companies = ['SAP', 'Datev', 'Apple', 'Microsoft', 'IBM', ]

colors = ['c', 'y', 'k', 'r', 'm']

# below we plot the chart, first we put in the 'slices' array, 
# label it with the companies names and put in the colors.  
# Startangle specifies the starangle of the graph. 
# Here we choose a 90 degree angle, 
# which means that the first division will be a vertical line. 
# Then we specify a shadow (Optional) and with explode we can 
# pull out a slice a bit (here: Datev).
# autopct displays the percentages
plt.pie(slices, labels = companies, colors = colors, 
        startangle = 90, shadow = True, 
        explode = (0, 0.1, 0, 0, 0), autopct = '%1.1f%%')

plt.show()
```

# How to deal with multiple plots

The following code samples will explain to you how can plot multiple plots in different ways. First of all we need numpy to do this, since we have to use numpy arrays. 

```python
import numpy as np

def function(parameter):
    return np.exp(-parameter)*np.cos(2*np.pi*parameter)

# Here we create 2 numpy arrays. Let me explain how this 
# works with the first array as an example:
# The first goes from 0.0 to 5.0 in the steps of 0.1
parameter1 = np.arange(0.0, 5.0, 0.1)
parameter2 = np.arange(0.0, 5.0, 0.02)

# Let me tell you what subplot does: With 211 we tell matplotlib 
# that there will be 2 plots vertically, 1 horizontally and this 
# plot here will be the first one.   
plt.subplot(211)
plt.plot(parameter1, function(parameter1), 'bo', parameter2, 
         function(parameter2))

# Here we tell subplot again that there will be 2 plots vertically, 
# 1 horizontally and that this one will be the second one.
plt.subplot(212)
plt.plot(parameter2, np.cos(2*np.pi*parameter2))
plt.show()
```

```python
# To illustrate better what subplot() does, I will 
# change the values we put in it below, 
# so that the graphs are alligned horizontally.
plt.subplot(221)
plt.plot(parameter1, function(parameter1), 'bo', 
         parameter2, function(parameter2))

plt.subplot(222)
plt.plot(parameter2, np.cos(2*np.pi*parameter2))
plt.show()
```
